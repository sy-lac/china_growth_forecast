# data preparation

```python
x = data2.drop('Gross Domestic Product Current Quarter(100 million yuan)', axis=1)

A = []

for i in range(12, len(x)) :
  a = x.iloc[i-12: i]
  A.append(a)

X = np.array(A)
X.shape
```

```python
Y = data2['Gross Domestic Product Current Quarter(100 million yuan)']

B = []

for i in range(12, len(Y)) :
  b = Y.iloc[i]
  B.append(b)

y = np.array(B)
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, random_state=0, shuffle=False)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_sc = scaler.fit_transform(X_train.reshape(189, 36))
X_test_sc = scaler.transform(X_test.reshape(21, 36))

X_train_sc = X_train_sc.reshape(189, 12, 3)
X_test_sc = X_test_sc.reshape(21, 12, 3)
y_train = y_train.reshape(189, 1)
y_test = y_test.reshape(21, 1)

X_train_sc.shape, X_test_sc.shape, y_train.shape, y_test.shape
```

# RNN model

```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense, Embedding
import tensorflow as tf

np.random.seed(42)
tf.random.set_seed(42)

def model() :
  model = Sequential()
  model.add(SimpleRNN(units=8, return_sequences=True, kernel_initializer='random_normal'))
  model.add(SimpleRNN(units=16, return_sequences=True, kernel_initializer='random_normal'))
  model.add(SimpleRNN(units=8, return_sequences=False, kernel_initializer='random_normal'))
  model.add(Dense(units=1, activation='linear', kernel_initializer='random_normal'))
  return model
```

```python
mdl = model()
mdl.compile(optimizer='rmsprop', loss='mean_squared_error')

mdl.fit(x=X_train_sc, y=y_train, validation_data=(X_test_sc, y_test), epochs=500, batch_size=12)
```

```python
mdl.summary()
```

```python
plt.plot(np.arange(len(X_train_sc)), y_train, label='Train')
plt.plot(np.arange(len(X_train_sc), len(X_train_sc)+len(X_test_sc), 1), y_test, label='Test')
plt.plot(np.arange(len(X_train_sc), len(X_train_sc)+len(X_test_sc), 1), mdl.predict(X_test_sc), label='Test prediction')
plt.legend()
plt.show()
```

# lstm model

```python
import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import GRU, Input, Dense, TimeDistributed, Activation, RepeatVector, Bidirectional, Dropout, LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import sparse_categorical_crossentropy
```

```python
def chinese_gdp_model(input_shape, output_dim):

  model = Sequential()
  model.add(LSTM(units=64, input_shape=input_shape, return_sequences=True))
  model.add(Dropout(0.2))
  model.add(LSTM(units=64, return_sequences=False))
  model.add(Dense(units=output_dim, activation='linear'))

  return model
```

```python
chinese_gdp = chinese_gdp_model((12, 3), 1)
chinese_gdp.compile(optimizer=Adam(learning_rate = 0.005), loss='mean_squared_error')

chinese_gdp.fit(X_train_sc, y_train, validation_data=(X_test_sc, y_test), epochs=500, batch_size=1)
```

```python
plt.plot(np.arange(len(X_train_sc)), y_train, label='Train')
plt.plot(np.arange(len(X_train_sc), len(X_train_sc)+len(X_test_sc), 1), y_test, label='Test')
plt.plot(np.arange(len(X_train_sc), len(X_train_sc)+len(X_test_sc), 1), chinese_gdp.predict(X_test_sc), label='Test prediction')
plt.legend()
plt.show()
```

# linear regression model

```python
from sklearn.model_selection import train_test_split

S = data2.drop('Gross Domestic Product Current Quarter(100 million yuan)', axis=1)
t = data2['Gross Domestic Product Current Quarter(100 million yuan)']

S_train, S_test, t_train, t_test = train_test_split(S, t, train_size=0.9, random_state=0, shuffle=False)
S_train.shape, S_test.shape, t_train.shape, t_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(S_train)

S_train_sc = scaler.transform(S_train)
S_test_sc = scaler.transform(S_test)
```

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()

lr.fit(S_train_sc, t_train)
```

```python
t_pred_train = lr.predict(S_train_sc)
t_pred_test = lr.predict(S_test_sc)
```

```python
from sklearn.metrics import mean_squared_error

mse1 = mean_squared_error(t_pred_train, t_train)
mse2 = mean_squared_error(t_pred_test, t_test)

print(np.sqrt(mse1)), print(np.sqrt(mse2))
```

```python
plt.plot(np.arange(len(S_train_sc)), t_train, label='Train')
plt.plot(np.arange(len(S_train_sc), len(S_train_sc)+len(S_test_sc), 1), t_test, label='Test')
plt.plot(np.arange(len(S_train_sc), len(S_train_sc)+len(S_test_sc), 1), lr.predict(S_test_sc), label='Test prediction')
plt.legend()
plt.show()
```
